#### Loding the data - trader behavior

```python
import pandas as pd

trades = pd.read_csv("historical_data.csv")
sentiment = pd.read_csv("fear_greed_index.csv")
```

#### Convert it to datetime.

```python
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], dayfirst=True)
```

```python
trades.head(5)
```

#### extract date only

```python
trades['date'] = trades['Timestamp IST'].dt.date
```

```python
trades.head(5)
```

#### Convert Fear/Greed Date

```python
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
```

#### Merge Both Datasets

```python
merged = pd.merge(trades, sentiment[['date','classification']], on='date', how='left')
```

```python
merged.head(5)
```

```python
merged['classification'].value_counts()
```

#### Check Missing Sentiment Values

```python
merged['classification'].isnull().sum()
```

```python
merged = merged.dropna(subset=['classification'])
```

#### Profit during Fear vs Greed

```python
merged.groupby('classification')['Closed PnL'].mean()
```

#### Total Profit


```python
merged.groupby('classification')['Closed PnL'].sum()
```

#### Average Leverage Behavior

```python
merged.groupby('classification')['Size USD'].mean()
```

#### Best Traders

```python
top_traders = merged.groupby('Account')['Closed PnL'].sum().sort_values(ascending=False)
print(top_traders.head(10))
```

#### Trading Volume

```python
merged.groupby('classification')['Size USD'].sum()
```

#### Profit vs Market Sentiment

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='classification', y='Closed PnL', data=merged)
plt.show()
```

#### total profit


```python
profit_sentiment = merged.groupby('classification')['Closed PnL'].sum()

profit_sentiment.plot(kind='bar', figsize=(8,5))
plt.title("Total Profit by Market Sentiment")
plt.ylabel("Total Profit")
plt.show()
```

#### trading values by sentiments


```python
volume = merged.groupby('classification')['Size USD'].sum()

volume.plot(kind='bar', figsize=(8,5))
plt.title("Trading Volume During Fear vs Greed")
plt.ylabel("Total Trade Volume")
plt.show()
```

```python
sns.countplot(x='classification', hue='Side', data=merged)
plt.title("Buy vs Sell Activity in Fear vs Greed Markets")
plt.show()
```

```python
sns.scatterplot(x='Fee', y='Closed PnL', data=merged)
plt.title("Fee vs Profit Relationship")
plt.show()
```

```python
top_traders = merged.groupby('Account')['Closed PnL'].sum().sort_values(ascending=False).head(10)

top_traders.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Profitable Traders")
plt.ylabel("Total Profit")
plt.show()
```

```python
sentiment_count = merged.groupby(['date','classification']).size().unstack()

sentiment_count.plot(figsize=(12,6))
plt.title("Market Sentiment Trend Over Time")
plt.show()
```
